```python
import pandas as data_reader
import numpy as np
data = data_reader.read_csv("./adult.csv",1,",")
data_array = [data]
```

```python
data['country'] = data['country'].replace(' ?',np.nan)
data['workclass'] = data['workclass'].replace(' ?',np.nan)
data['occupation'] = data['occupation'].replace(' ?',np.nan)
data['sex'] = data['sex'].replace(' ?',np.nan)

data.dropna(how='any',inplace=True)
```

```python
data['salary']=data['salary'].map({' <=50K':1, ' >50K':0}).astype(int)
data['sex'] = data['sex'].map({' Male':1,' Female':0}).astype(int)

for dataset in data_array:
    dataset.loc[dataset['country'] != ' United-States', 'country'] = 'Non-US'
    dataset.loc[dataset['country'] == ' United-States', 'country'] = 'US'
    
data['country'] = data['country'].map({'US':1,'Non-US':0}).astype(int)

data['marital-status'] = data['marital-status']
.replace([' Divorced',' Married-spouse-absent',' Never-married',' Separated',' Widowed'],'Single')
data['marital-status'] = data['marital-status']
.replace([' Married-AF-spouse',' Married-civ-spouse'],'Couple')

data['marital-status'] = data['marital-status'].map({'Couple':0,'Single':1})

data['relationship'] = data['relationship']
.map({' Unmarried':0,' Wife':1,' Husband':2,' Not-in-family':3,' Own-child':4,' Other-relative':5})

data['race']= data['race'].map({' White':0,' Amer-Indian-Eskimo':1,
                                ' Asian-Pac-Islander':2,' Black':3,' Other':4})
   
data['workclass']= data['workclass'].map({' Federal-gov':0,' Local-gov':1,
                                          ' State-gov':2,' Private':3,' Self-emp-inc':4,' Self-emp-not-inc':5})    
    
data.drop(labels=['education','occupation', 'fnlwgt'],axis=1,inplace=True)
    
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
data.dropna(how='any',inplace=True)
X=data.drop(['salary'],axis=1)
y=data['salary']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5,random_state=22)
print ("Train dataset: {0}{1}".format(X_train.shape, y_train.shape))
print ("Test dataset: {0}{1}".format(X_test.shape, y_test.shape))

```

```python
print ("Train dataset: {0}{1}".format(X_train.shape, y_train.shape))
print ("Test dataset: {0}{1}".format(X_test.shape, y_test.shape))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn import model_selection
from sklearn.metrics import accuracy_score

mlp = (MLPClassifier(solver='adam', hidden_layer_sizes=(40, 40), max_iter=2000,
activation='relu', learning_rate_init=0.02))

mlp.fit(X_train, y_train)
prediction = mlp.predict(X_test)
acc_score = accuracy_score(y_test, prediction)
print ('{0}: {1}'.format('Neural network',acc_score))
```

```python
models = []
names = ['Random Forest','GaussianNB','DecisionTreeClassifier', 'KNeighborsClassifier']

models.append((RandomForestClassifier(n_estimators=10)))
models.append((GaussianNB()))
models.append((DecisionTreeClassifier()))
models.append((KNeighborsClassifier()))

for i in range(0,len(models)):    
    score=models[i].fit(X_train,y_train)
    prediction = models[i].predict(X_test)
    acc_score = accuracy_score(y_test,prediction)     
    print ('-'*40)
    print ('{0}: {1}'.format(names[i],acc_score))
```
